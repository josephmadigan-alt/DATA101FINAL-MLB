```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
pio.renderers.defaule = 'colab'

from itables import show
```

```python
df25 = pd.read_csv('25mlb-player-stats-Batters.csv')
df24 = pd.read_csv('24mlb-player-stats-Batters.csv')
df23 = pd.read_csv('23mlb-player-stats-Batters.csv')
df22 = pd.read_csv('22mlb-player-stats-Batters.csv')
df21 = pd.read_csv('21mlb-player-stats-Batters.csv')
df20 = pd.read_csv('20mlb-player-stats-Batters.csv')
df19 = pd.read_csv('19mlb-player-stats-Batters.csv')
df18 = pd.read_csv('18mlb-player-stats-Batters.csv')
df17 = pd.read_csv('17mlb-player-stats-Batters.csv')
df16 = pd.read_csv('16mlb-player-stats-Batters.csv')
df15 = pd.read_csv('15mlb-player-stats-Batters.csv')
df14 = pd.read_csv('14mlb-player-stats-Batters.csv')
df13 = pd.read_csv('13mlb-player-stats-Batters.csv')
df12 = pd.read_csv('12mlb-player-stats-Batters.csv')
df11 = pd.read_csv('11mlb-player-stats-Batters.csv')
df10 = pd.read_csv('10mlb-player-stats-Batters.csv')
```

```python
df25['Year'] = '2025'
df24['Year'] = '2024'
df23['Year'] = '2023'
df22['Year'] = '2022'
df21['Year'] = '2021'
df20['Year'] = '2020'
df19['Year'] = '2019'
df18['Year'] = '2018'
df17['Year'] = '2017'
df16['Year'] = '2016'
df15['Year'] = '2015'
df14['Year'] = '2014'
df13['Year'] = '2013'
df12['Year'] = '2012'
df11['Year'] = '2011'
df10['Year'] = '2010'
```

```python
df_main = pd.concat([df25,df24,df23,df22,df21,df20,df19,df18,df17,df16,df15,df14,df13,df12,df11,df10])
show(df_main)
```

```python
df_main = df_main.reset_index()
```

```python
NL_teams = ['ARI','ATL','CHC','CIN','COL','LAD','MIA','MIL','NYM','PHI','PIT','SD','SF','STL','WAS']
df_main['League'] = 'no info'
for k in df_main.index:
    if df_main['Team'].iloc[k] in NL_teams:
        df_main.loc[k,'League'] = 'NL'
    else:
        df_main.loc[k,'League'] = 'AL'
show(df_main)
```

```python
df_main['adj_age'] = 'none'
for k in df_main.index:
    df_main.loc[k,'adj_age'] = df_main['Age'].iloc[k] - (2025 - int(df_main['Year'].iloc[k]))

show(df_main)
```

```python
franmask = df_main['Player'] == 'Francisco Lindor'

show(df_main[franmask])
```

```python
df_main.to_csv('output.csv', index=False) 
```
