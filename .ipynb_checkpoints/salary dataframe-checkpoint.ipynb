```python
try:
    DF = pd.read_csv('mlb_salary_data.csv', encoding='latin1')
except UnicodeDecodeError:
    try:
        DF = pd.read_csv('mlb_salary_data.csv', encoding='cp1252')
    except UnicodeDecodeError:
        DF = pd.read_csv('mlb_salary_data.csv', encoding='iso-8859-1')
```
