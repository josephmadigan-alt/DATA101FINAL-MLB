```python
df15 = pd.read_csv('./sprint_speed(2015).csv')
df25 = pd.read_csv('./sprint_speed(2025).csv')
df15.head()
```

```python
fig = px.scatter(
    df15,
    x="age",
    y="sprint_speed",
    hover_data=["last_name, first_name", "team", "position"],
    title="Age vs Sprint Speed",
    labels={
        "age": "Player Age",
        "sprint_speed": "Sprint Speed (ft/sec)"
    }
)

fig.update_traces(marker=dict(size=8, opacity=0.7))
fig.update_layout(
    template="plotly_white"
)

fig.show()
```

```python
fig = px.scatter(
    df15,
    x="age",
    y="sprint_speed",
    color="position",
    hover_data=["last_name, first_name", "team"],
    title="Age vs Sprint Speed by Position (2015)"
)

fig.show()
```

```python
fig = px.scatter(
    df25,
    x="age",
    y="sprint_speed",
    color="position",
    hover_data=["last_name, first_name", "team"],
    title="Age vs Sprint Speed by Position (2025)"
)

fig.show()
```

```python
fig = px.scatter(
    df15,
    x="age",
    y="sprint_speed",
    trendline="ols",
    hover_data=["last_name, first_name", "team"],
    title="Age vs Sprint Speed with Trendline (2015)"
)

fig.show()

```

```python
fig = px.scatter(
    df25,
    x="age",
    y="sprint_speed",
    trendline="ols",
    hover_data=["last_name, first_name", "team"],
    title="Age vs Sprint Speed with Trendline (2025)"
)

fig.show()
```

```python
fig = px.scatter(
    df15,
    x="age",
    y="sprint_speed",
    size="competitive_runs",
    color="position",
    hover_data=["last_name, first_name", "team"],
    title="Age vs Sprint Speed (Sized by Competitive Runs) (2015)"
)

fig.show()
```

```python
fig = px.scatter(
    df25,
    x="age",
    y="sprint_speed",
    color = 'position',
    size="competitive_runs",
    hover_data=["last_name, first_name", "team"],
    title="Age vs Sprint Speed (Sized by Competitive Runs) (2025)"
)

fig.show()
```
