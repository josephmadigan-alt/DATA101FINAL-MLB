```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
pio.renderers.defaule = 'colab'

from itables import show
```

```python
DF = pd.read_csv('contracts_2025.csv')
show(DF)
```

```python
columns = ['Name','Salary']
df = DF[columns].copy()
show(df)
```

```python
df['Salary'] = df['Salary'].apply(lambda x: x.replace('$',''))
df['Salary'] = df['Salary'].apply(lambda x: x.replace(',',''))
df['Salary'] = df['Salary'].apply(lambda x: float(x))
show(df)
```

```python
df['salcat'] = 'none'
for k in df.index:
    if df['Salary'].loc[k] < 1000000:
        df.loc[k, 'salcat'] = '<1M'
    elif (df['Salary'].loc[k] >= 1000000) & (df['Salary'].loc[k] < 10000000):
        df.loc[k, 'salcat'] = '1 - 10M'
    elif (df['Salary'].loc[k] >= 10000000) & (df['Salary'].loc[k] < 20000000):
        df.loc[k, 'salcat'] = '10 - 20M'
    elif (df['Salary'].loc[k] >= 20000000) & (df['Salary'].loc[k] < 30000000):
        df.loc[k, 'salcat'] = '20 - 30M'
    elif (df['Salary'].loc[k] >= 30000000) & (df['Salary'].loc[k] < 40000000):
        df.loc[k, 'salcat'] = '30 - 40M'
    elif (df['Salary'].loc[k] >= 40000000) & (df['Salary'].loc[k] < 50000000):
        df.loc[k, 'salcat'] = '40 - 50M'
    else:
        df.loc[k, 'salcat'] = '>50M'
```

```python
show(df)
```

```python
df.to_csv('2025_contract_avg.csv', index = False)
```
