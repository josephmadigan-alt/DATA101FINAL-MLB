```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
pio.renderers.default = 'colab'

from itables import show
import requests
from bs4 import BeautifulSoup
```

```python
website = 'https://www.fangraphs.com/leaders/major-league?type=23&pagenum=1&pageitems=2000000000&sortcol=11&sortdir=default'
```

```python
raw_code = requests.get(website)
html_doc = raw_code.text
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
soup.find_all?
```

```python
result = soup.find_all('td', class_ = 'align-right', attrs={"data-col-id": 'OBP+'})

df = pd.DataFrame()
df['OBP+'] = result
df['OBP+'] = df['OBP+'].apply(lambda x: float(x.text))
show(df)
```

```python
result = soup.find_all('td', class_ = 'align-right', attrs={"data-col-id": 'Name'})
df = pd.DataFrame()
df['Name'] = result
df['Name'] = df['Name'].apply(lambda x: x.text)

result = soup.find_all('td', class_ = 'align-right', attrs={"data-col-id": 'OBP+'})
df['OBP+'] = result
df['OBP+'] = df['OBP+'].apply(lambda x: float(x.text))

result = soup.find_all('td', class_ = 'align-right', attrs={"data-col-id": 'SLG+'})
df['SLG+'] = result
df['SLG+'] = df['SLG+'].apply(lambda x: float(x.text))
show(df)
```
