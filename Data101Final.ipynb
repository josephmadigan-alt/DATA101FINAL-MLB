```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
pio.renderers.defaule = 'colab'

from itables import show

# This stops a few warning messages from showing
pd.options.mode.chained_assignment = None
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Machine Learning Packages
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
```

```python
file_location = 'output.csv'
DF_raw = pd.read_csv(file_location)
```

```python
DF_raw['PA'] = DF_raw['AB'] + DF_raw['BB'] + DF_raw['HBP']
```

```python
#minimun 300 Plate Appearances
df = DF_raw[DF_raw['PA'] >= 300]
```

```python
First = df[df['Pos'] == '1B']
Second = df[df['Pos'] == '2B']
Third = df[df['Pos'] == '3B']
Short = df[df['Pos'] == 'SS']
Outfield = df[df['Pos'] == 'OF']
Catcher = df[df['Pos'] == 'C']
```

```python
First_avg = np.mean(First['AVG'])
Second_avg = np.mean(Second['AVG'])
Third_avg = np.mean(Third['AVG'])
Short_avg = np.mean(Short['AVG'])
Outfield_avg = np.mean(Outfield['AVG'])
Catcher_avg = np.mean(Catcher['AVG'])

print('1B-', First_avg)
print('2B-', Second_avg)
print('3B-', Third_avg)
print('SS-', Short_avg)
print('OF-', Outfield_avg)
print('C-', Catcher_avg)
```

```python
First_OBP = np.mean(First['OBP'])
Second_OBP = np.mean(Second['OBP'])
Third_OBP = np.mean(Third['OBP'])
Short_OBP = np.mean(Short['OBP'])
Outfield_OBP = np.mean(Outfield['OBP'])
Catcher_OBP = np.mean(Catcher['OBP'])

print('1B-', First_OBP)
print('2B-', Second_OBP)
print('3B-', Third_OBP)
print('SS-', Short_OBP)
print('OF-', Outfield_OBP)
print('C-', Catcher_OBP)
```

```python
First_SLG = np.mean(First['SLG'])
Second_SLG = np.mean(Second['SLG'])
Third_SLG = np.mean(Third['SLG'])
Short_SLG = np.mean(Short['SLG'])
Outfield_SLG = np.mean(Outfield['SLG'])
Catcher_SLG = np.mean(Catcher['SLG'])

print('1B-', First_SLG)
print('2B-', Second_SLG)
print('3B-', Third_SLG)
print('SS-', Short_SLG)
print('OF-', Outfield_SLG)
print('C-', Catcher_SLG)
```

```python
First_OPS = np.mean(First['OPS'])
Second_OPS = np.mean(Second['OPS'])
Third_OPS = np.mean(Third['OPS'])
Short_OPS = np.mean(Short['OPS'])
Outfield_OPS = np.mean(Outfield['OPS'])
Catcher_OPS = np.mean(Catcher['OPS'])

print('1B-', First_OPS)
print('2B-', Second_OPS)
print('3B-', Third_OPS)
print('SS-', Short_OPS)
print('OF-', Outfield_OPS)
print('C-', Catcher_OPS)
```

```python
positions = ['1B', '2B', '3B', 'SS', 'OF', 'C']
ops_values = [First_OPS, Second_OPS, Third_OPS, Short_OPS, Outfield_OPS, Catcher_OPS]

plt.figure(figsize=(8,5))
plt.bar(positions, ops_values, color='skyblue')
plt.title('Average OPS by Position')
plt.xlabel('Position')
plt.ylabel('OPS')

plt.ylim(min(ops_values) * 0.95, max(ops_values) * 1.05)

plt.show()
```

```python
avg_values = [First_avg, Second_avg, Third_avg, Short_avg, Outfield_avg, Catcher_avg]

plt.figure(figsize=(8,5))
plt.bar(positions, avg_values, color='lightgreen')
plt.title('Average AVG by Position')
plt.xlabel('Position')
plt.ylabel('AVG')
plt.ylim(min(avg_values) * 0.95, max(avg_values) * 1.05)
plt.show()
```

```python
slg_values = [First_SLG, Second_SLG, Third_SLG, Short_SLG, Outfield_SLG, Catcher_SLG]

plt.figure(figsize=(8,5))
plt.bar(positions, slg_values, color='salmon')
plt.title('Average SLG by Position')
plt.xlabel('Position')
plt.ylabel('SLG')
plt.ylim(min(slg_values) * 0.95, max(slg_values) * 1.05)
plt.show()
```

```python
# Compute XBH and XBH%
df['XBH'] = df['2B'] + df['3B'] + df['HR']
df['XBH_pct'] = df['XBH'] / df['H']

First = df[df['Pos'] == '1B']
Second = df[df['Pos'] == '2B']
Third = df[df['Pos'] == '3B']
Short = df[df['Pos'] == 'SS']
Outfield = df[df['Pos'] == 'OF']
Catcher = df[df['Pos'] == 'C']

# Calculate the mean XBH% for each position
pos_data = {
    '1B': First['XBH_pct'].mean(),
    '2B': Second['XBH_pct'].mean(),
    '3B': Third['XBH_pct'].mean(),
    'SS': Short['XBH_pct'].mean(),
    'OF': Outfield['XBH_pct'].mean(),
    'C': Catcher['XBH_pct'].mean(),
}

# Convert to DataFrame for plotting
pos_df = pd.DataFrame.from_dict(pos_data, orient='index', columns=['XBH_pct'])

# Plot
plt.figure(figsize=(8,6))
plt.bar(pos_df.index, pos_df['XBH_pct'])
plt.xlabel("Position")
plt.ylabel("XBH%")
plt.ylim(pos_df['XBH_pct'].min() * 0.95, pos_df['XBH_pct'].max() * 1.05)  # zoom in
plt.title("Extra-Base-Hit Percentage by Position")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
